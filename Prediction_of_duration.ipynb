```python
# Assuming x_train (features) and y_train (target) are already defined as pandas DataFrame and Series
print(x_train.shape)  # Check the shape of x_train to confirm the number of features

# Assuming x_train has 5 features, set column names accordingly
x_train_df = pd.DataFrame(x_train, columns=['Days for shipment (scheduled)', 'Benefit per order', 
                                            'Sales per customer', 'Category Name', 'Customer Zipcode'])

# Train the Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Get the intercept and coefficients
intercept = model.intercept_
coefficients = model.coef_

# Create the equation
equation = f"y = {intercept:.4f}"
for coef, feature in zip(coefficients, x_train_df.columns):  # For DataFrame, use column names
    equation += f" + ({coef:.4f} * {feature})"

print(f"Regression Equation: {equation}")
```
